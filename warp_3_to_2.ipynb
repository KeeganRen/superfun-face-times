```python
def makeRotation(a,b,c):
    # skew symmetric... exponentiated to get rotation magic
    ss = array([[0,a,b],[-a,0,c],[-b,-c,0]])
    r = scipy.linalg.block_diag(scipy.linalg.expm(ss), [[1]])
    return matrix(r)

def makeTranslation(tx, ty):
    m = eye(4)
    m[3,0] = tx
    m[3,1] = ty
    return matrix(m)

def makeScale(s):
    m = diag([s, s, s, 1])
    return matrix(m)
    

def makeTransformation(a,b,c,tx,ty,s):
    return  makeScale(s) * makeRotation(a,b,c) * makeTranslation(tx,ty)

def residuals(p):
    m = makeTransformation(*p)
    aug_3d = vstack((t_3d.T, ones(27))).T
    transformed = asarray(aug_3d*m)
    return (transformed[:, 0:2] - t_2d).ravel()
   
    
def doTransform(p):
    m = makeTransformation(*p)
    aug_3d = vstack((t_3d.T, ones(27))).T
    transformed = asarray(aug_3d*m)
    return transformed[:, 0:3]


soln,res = scipy.optimize.leastsq(residuals, [0,0,0,0,0,0])
print "solution", soln
print "res", res
```

```python
best_transformation = makeTransformation(*soln)

print best_transformation

rec_3d = doTransform(soln)

scatter(rec_3d[:,0], -rec_3d[:,2], color="green")
#scatter(t_2d[:,0], -t_2d[:,1], color="red")
```
